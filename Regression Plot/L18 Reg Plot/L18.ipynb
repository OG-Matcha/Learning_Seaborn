### Import library

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Read data

```python
tips = sns.load_dataset('tips')
tips.head()
```

### Creat a basic reg plot

```python
sns.regplot(data=tips, x='total_bill', y='tip')
```

### Change the styling of reg plot, only line, only scatter, show/hide the ci, change the ci value, change n_boot

```python
sns.regplot(data=tips, x='total_bill', y='tip'
           , color='red'
           , marker='+')
```

```python
sns.regplot(data=tips, x='total_bill', y='tip'
        #    , color='red'
        #    , marker='+'
           , line_kws=dict(color='red', linestyle='--')
           , scatter_kws=dict(s=100, color='green', alpha=0.5)
           , ci=60
           , n_boot=500)
```
