### Import library

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### Read data

```python
mart = pd.read_excel("../../Data/supermarket_sales.xlsx")
mart = mart[['Gender','Payment','Unit price','Quantity','Total','gross income','Branch']]

mart.head(10)
```

### Including/Combining a KDE Plot with Histogram using Dis Plot

```python
sns.displot(data=mart, x='Total', kde=True)
```

### Including/Combining a Rug Plot with Histogram & KDE using Dis Plot | Creating a Histogram, KDE and Rug plot altogether in one plot using Dis Plot

```python
sns.displot(data=mart, x='Total', kde=True, rug=True, rug_kws={'height':0.05}, kde_kws={'bw_adjust':1})
```

### Grouping a Dis Plot basis on a categorical variable

```python
sns.displot(data=mart, x='Total', kde=True, rug=True, rug_kws={'height':0.05}, hue='Payment', multiple='stack')
```

### Separating a Displot in multiple columns basis on the category in supplied categorical variable

```python
sns.displot(data=mart, x='Total', kde=True, rug=True, rug_kws={'height':0.05}, hue='Payment', multiple='stack', col='Gender')
```

### Separating a Displot in multiple rows basis on the category in supplied categorical variable

```python
sns.displot(data=mart, x='Total', kde=True, rug=True, rug_kws={'height':0.05}, hue='Payment', multiple='stack', col='Branch', row='Gender')
```

### Creating Bar Plots using Dis plot in Seaborn

```python
sns.displot(data=mart, y='Total', kde=True, rug=True, rug_kws={'height':0.05}, hue='Payment', multiple='stack', col='Branch', row='Gender')
```

### Creating ECDF Plot using Displot in Seaborn

```python
sns.displot(data=mart, y='Total', kind='ecdf', hue='Payment', col='Branch', row='Gender')
```
