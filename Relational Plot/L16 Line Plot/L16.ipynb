### Import library

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Read data

```python
mart = pd.read_excel("../../Data/mart_train_sample.xlsx").sample(50)

mart.head()
```

### Creat a basic line plot and try different estimators

```python
sns.lineplot(data=mart, x="Outlet_Year", y="Sales", errorbar=None, estimator=None)
```

### Test with different Confidence Intervals and with different number of Bootstraping

```python
sns.lineplot(data=mart, x="Outlet_Year", y="Sales", n_boot=2)
```

### Grouping using HUE and use different Palettes

```python
sns.lineplot(data=mart, x="Outlet_Year", y="Sales", errorbar=None
            , hue='Outlet_Size'
            , palette=['red', 'green', 'blue'])
```

### Grouping using Style, use different Marker and Dashes

```python
sns.lineplot(data=mart, x="Outlet_Year", y="Sales", errorbar=None
            , style='Outlet_Size'
            , markers=True
            , dashes=False)
```

### Grouping using Size

```python
sns.lineplot(data=mart, x="Outlet_Year", y="Sales", errorbar=None
            , size='Outlet_Size'
            , sizes=(0.5, 5))
```

### Use different semantic styling parameters on same or different variables

```python
sns.lineplot(data=mart, x="Outlet_Year", y="Sales", errorbar=None
            , hue='Outlet_Size'
            # , style='Outlet_Size'
            # , markers=True
             , size='Tier'
            )
```
