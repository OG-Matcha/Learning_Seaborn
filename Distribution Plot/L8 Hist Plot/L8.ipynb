### Import library

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statistics as sts
```

### Read data

```python
mart = pd.read_excel("../../Data/supermarket_sales.xlsx")
mart = mart[['Gender','Payment','Unit price','Quantity','Total','gross income']]

mart.head(10)
```

### Create a basic HISTOGRAM on X or Y axis

```python
sns.histplot(data=mart, y='Total')
```

### Change the Bins Width using binwidth arguement

```python
sns.histplot(data=mart, x='Total', binwidth=150)
```

### Change Number of Bins and Interval

```python
plt.figure(figsize=(15,5))
plt.xticks(np.arange(0, 1100, 50))

sns.histplot(data=mart, x='Total', bins=np.arange(0, 1100, 50))
```

### Combine with KDE

```python
sns.histplot(data=mart, x='Total', kde=True)
```

### Use a categorical variable in HUE and STACK it using MULTIPLE agruement

```python
sns.histplot(data=mart, x='Total', hue='Payment', multiple='stack')
```

### Make it a step/poly plot using ELEMENT arguement and change the FILL

```python
sns.histplot(data=mart, x='Total', hue='Payment', element='poly', fill=False)
```

### Use categorical variable and shrink it

```python
sns.histplot(data=mart, x='Total', shrink=0.7, stat='probability')
```

### Create a Bivariate Histogram

```python
sns.histplot(data=mart, x='Total', y='gross income')
```
