### Import library

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### Read data

```python
mart = pd.read_excel("../../Data/supermarket_sales.xlsx")

mart.head()
```

### Create a basic BOX plot on one NUMERIC variable

```python
sns.boxplot(y="Total", data=mart, width=0.2)
```

### Create a basic BOX plot on one numeric variable by a CATEGORICAL variable

```python
sns.boxplot(x="Payment", y="Total", data=mart)
```

### Create a basic BOX plot on one numeric variable by TWO CATEGORICAL variable using HUE attribute

```python
sns.boxplot(x="Payment", y="Total", data=mart, hue="Gender")
```

### Add MEAN marker in the box plot using Showmeans attribute and change its style using meanprops

```python
sns.boxplot(x="Payment", y="Total", data=mart, hue="Gender", showmeans=True, meanprops={"marker":"o"
                                                                                        ,"markerfacecolor":"white"
                                                                                        ,"markersize":"10"
                                                                                        ,"markeredgecolor":"black"})
```

### Make HORIZONTAL box plot

```python
sns.boxplot(x="Total", y="Payment", data=mart, hue="Gender", showmeans=True, meanprops={"marker":"o"
                                                                                        ,"markerfacecolor":"white"
                                                                                        ,"markersize":"10"
                                                                                        ,"markeredgecolor":"black"})
```

### Change PALETTE, LINE WIDTH etc..

```python
sns.boxplot(x="Payment", y="Total", data=mart, hue="Gender", showmeans=True, meanprops={"marker":"o"
                                                                                        ,"markerfacecolor":"white"
                                                                                        ,"markersize":"10"
                                                                                        ,"markeredgecolor":"black"}
            , palette="CMRmap", linewidth=5)
```

### Create box plot for EACH OF THE NUMERIC VARIABLE in the dataframe

```python
plt.figure(figsize=(15, 5))

sns.boxplot(data=mart)
```
