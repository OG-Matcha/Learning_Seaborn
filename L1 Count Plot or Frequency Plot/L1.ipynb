### import library

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### Read data

```python
mart = pd.read_excel("../Data/supermarket_sales.xlsx")

mart.head()
```

### Create a basic COUNTPLOT to get number of transaction for each of the product line and change the figure size

```python
plt.figure(figsize=(15, 5))

sns.countplot(x="Product line", data=mart)
```

### Make it a horizontal bar plot

```python
sns.countplot(y="Product line", data=mart)
```

### Add hue to get the count on two categories i.e. Product line and Gender

```python
plt.figure(figsize=(15, 5))

sns.countplot(x="Product line", hue="Gender", data=mart)
```

### Using different color palette

```python
plt.figure(figsize=(15, 5))

sns.countplot(x="Product line", hue="Gender", data=mart, palette="Wistia")
```

### Change style using facecolor, linewidth and edge color

```python
plt.figure(figsize=(15, 5))

sns.countplot(x="Product line", data=mart
              ,facecolor=(1, 0, 1, 0)
              ,linewidth=5
              ,edgecolor=sns.color_palette("dark", 3))
```
