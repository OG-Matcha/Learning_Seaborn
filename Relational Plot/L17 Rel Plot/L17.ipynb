### Import library

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Read data

```python
mart = pd.read_excel("../../Data/mart_linePlot.xlsx").sample(n=50)
mart.columns = mart.columns.str.lower()

mart.to_excel("../../Data/mart_relPlot.xlsx", index=False)
```

### Creating the first REL PLOT

```python
sns.relplot(data=mart, x='outlet_year', y='sales')
```

### Changing the kind to Line / Scatter Plot

```python
sns.relplot(data=mart, x='outlet_year', y='sales', kind='line', errorbar=None)
```

### Grouping based on a categorical variable using HUE arguement

```python
sns.relplot(data=mart, x='outlet_year', y='sales', s=100, hue='outlet_size')
```

### Creating a separate plot on column for each of the unique iteam in the supplied categorica variable

```python
sns.relplot(data=mart, x='outlet_year', y='sales', s=100, hue='outlet_size'
            , col='item_type')
```

### Using COL_WRAP to wrap the columns through different rows

```python
sns.relplot(data=mart, x='outlet_year', y='sales', s=100, hue='outlet_size'
            , col='item_type'
            , col_wrap=3)
```

```python
sns.relplot(data=mart, x='outlet_year', y='sales', s=100, hue='outlet_size'
            , col='outlet_size'
            , row='tier'
            , palette=['green','red','yellow'])
```
