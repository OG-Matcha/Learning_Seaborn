### Import library

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### Read data

```python
mart = pd.read_excel("../../Data/supermarket_sales.xlsx")
mart = mart[['Gender','Payment','Unit price','Quantity','Total','gross income']]

mart.head(10)
```

### Creating basic ECDF plot based on one numeric variable

```python
sns.ecdfplot(data=mart, x='gross income')
```

### Categorizing/Grouping the ECDF plot by a categorical variable value

```python
sns.ecdfplot(data=mart, x='gross income', hue='Gender')
```

### Showing count in ECDF plot|Switching the stat from Proportation to Count in a ECDF plot

```python
sns.ecdfplot(data=mart, y='gross income', hue='Gender', stat='count')
```
