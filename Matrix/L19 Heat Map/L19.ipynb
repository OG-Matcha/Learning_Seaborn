### Import library

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

### Read data

```python
mart = pd.read_excel("../../Data/mart_linePlot.xlsx", index_col=0)
mart.columns = mart.columns.str.lower()

mart.head(10)
```

### Creating a Pivot Table using a pandas dataframe for Heatmap

```python
martPiv = mart.pivot_table(index='outlet_year', columns='outlet_size', values='sales')
martPiv
```

### Creating a basic Heatmap

```python
sns.set_style('white')
plt.figure(figsize=(5, 5))

sns.heatmap(martPiv
            # , square=True
            )
```

### Showing/Annotating numbers on Heatmap

```python
sns.set_style('white')
plt.figure(figsize=(5, 5))

sns.heatmap(martPiv
#            , square=True
           , annot=True
            )
```

### Specifying decimal points for the annotated numbers on a Heatmap

```python
sns.set_style('white')
plt.figure(figsize=(5, 5))

sns.heatmap(martPiv
#            , square=True
           , annot=True
           , fmt='.0f'
            )
```

### Using annot key words to style the populted numbers on a Heatmap which includes - Font Size, Font Style etc..

```python
sns.set_style('white')
plt.figure(figsize=(5, 5))

sns.heatmap(martPiv
#            , square=True
           , annot=True
           , fmt='.0f'
           , annot_kws=dict(size=15, weight='bold')
            )
```

### Showing the lines on a Heatmap

```python
sns.set_style('white')
plt.figure(figsize=(5, 5))

sns.heatmap(martPiv
#            , square=True
           , annot=True
           , fmt='.0f'
           , annot_kws=dict(size=15, weight='bold')
           , linewidths=0.5
           , linecolor='black'
            )
```

### Using a different color palette in Heatmap

```python
sns.set_style('white')
plt.figure(figsize=(5, 5))

sns.heatmap(martPiv
#            , square=True
           , annot=True
           , fmt='.0f'
           , annot_kws=dict(size=15, weight='bold')
           , linewidths=0.5
        #    , linecolor='black'
           , cmap='OrRd_r'
           )
```

### Using Correlation matrix data to plot the Heatmap

```python
mart.corr()
```

```python
sns.heatmap(mart.corr())
```

### Adjusting minimum, maximum and center values on the Bar on a Heatmap

```python
sns.heatmap(mart.corr()
           , vmin=-1
           , vmax=1
           , center=0
           , cmap='OrRd_r'
           , annot=True
           , fmt='.1f'
           , annot_kws=dict(size=15, weight='bold')
           , linecolor='black'
           , linewidths=0.5)
```
