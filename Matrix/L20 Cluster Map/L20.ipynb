### Import library

```python
import pandas as pd
import matplotlib.pyplot as plt
# import numpy as np
import seaborn as sns
# from sunbird.categorical_encoding import frequency_encoding
```

### Read data

```python
mart = pd.read_excel("../../Data/mart_linePlot.xlsx", index_col=0)
mart.columns = mart.columns.str.lower()

mart.head(10)
```

```python
martPiv = mart.pivot_table(index='outlet_year', columns='outlet_size', values='sales')
martPiv.head()
```

###  Creating a basic Cluster Map

```python
sns.clustermap(martPiv)
```

### Having only Row Cluster in a Cluster Map

```python
sns.clustermap(martPiv
               , col_cluster=False)
```

### Having only Column Cluster in a Cluster Map

```python
sns.clustermap(martPiv
            #    , col_cluster=False
               , row_cluster=False
               )
```

### Styling in a Cluster Map

```python
sns.clustermap(martPiv
            #    , col_cluster=False
               , row_cluster=False
               , annot=True
               , fmt='.0f'
               )
```

### Changing Z Score value in a Cluster Map

```python
sns.clustermap(martPiv
            #    , col_cluster=False
               , row_cluster=False
               , annot=True
               , fmt='.0f'
               , z_score=1
               )
```

### Changing the Standard Scale Value in a Cluster Map

```python
sns.clustermap(martPiv
            #    , col_cluster=False
               , row_cluster=False
               , annot=True
               , fmt='.0f'
            #    , z_score=1
               , standard_scale=1
               )
```

### Line Styling in a Cluster Map

```python
sns.clustermap(martPiv
            #    , col_cluster=False
               , row_cluster=False
               , annot=True
               , fmt='.0f'
            #    , z_score=1
               , standard_scale=1
               , linewidth=0.5
               )
```
