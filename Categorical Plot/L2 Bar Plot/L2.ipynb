### Import library

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### Display the SEABORN dataset names and load one of them for example

```python
sns.get_dataset_names()

titanic = sns.load_dataset("titanic")

titanic.head()
```

### Read data

```python
mart = pd.read_excel("../../Data/supermarket_sales.xlsx")

mart.head()
```

### Create a basic BARPLOT

```python
plt.figure(figsize=(15, 5))

sns.barplot(x="Product line", y="Total", data=mart)
```

### Add HUE to the barplot

```python
plt.figure(figsize=(15, 5))

sns.barplot(x="Product line", y="Total", data=mart, hue="Gender")
```

### Make the barplot HORIZONTAL

```python
plt.figure(figsize=(15, 5))

sns.barplot(x="Total", y="Product line", data=mart)
```

### Plot the bars in a given ORDER

```python
x = mart["Product line"].sort_values()
x.unique()
```

```python
plt.figure(figsize=(15, 5))

sns.barplot(x="Product line", y="Total", data=mart, hue="Gender", order=['Electronic_accessories', 'Fashion_accessories',
       'Food_and_beverages', 'Health_and_beauty', 'Home_and_lifestyle',
       'Sports_and_travel'], hue_order=["Male", "Female"])
```

### Add CAP on the ERROR BAR

```python
plt.figure(figsize=(15, 5))

sns.barplot(x="Product line", y="Total", data=mart, hue="Gender", order=['Electronic_accessories', 'Fashion_accessories',
       'Food_and_beverages', 'Health_and_beauty', 'Home_and_lifestyle',
       'Sports_and_travel'], hue_order=["Male", "Female"], capsize=0.2)
```

### REMOVE the error bar using errorbar

```python
plt.figure(figsize=(15, 5))

sns.barplot(x="Product line", y="Total", data=mart, hue="Gender", order=['Electronic_accessories', 'Fashion_accessories',
       'Food_and_beverages', 'Health_and_beauty', 'Home_and_lifestyle',
       'Sports_and_travel'], errorbar=None)
```

### Change bar colors using COLOR attribute

```python
plt.figure(figsize=(15, 5))

sns.barplot(x="Product line", y="Total", data=mart, order=['Electronic_accessories', 'Fashion_accessories',
       'Food_and_beverages', 'Health_and_beauty', 'Home_and_lifestyle',
       'Sports_and_travel'], color='red')
```

### Change color using PALETTE attribute

```python
plt.figure(figsize=(15, 5))

sns.barplot(x="Product line", y="Total", data=mart, hue="Gender", order=['Electronic_accessories', 'Fashion_accessories',
       'Food_and_beverages', 'Health_and_beauty', 'Home_and_lifestyle',
       'Sports_and_travel'], palette="cividis")
```

### Using SATURATION parameter

```python
plt.figure(figsize=(15, 5))

sns.barplot(x="Product line", y="Total", data=mart, order=['Electronic_accessories', 'Fashion_accessories',
       'Food_and_beverages', 'Health_and_beauty', 'Home_and_lifestyle',
       'Sports_and_travel'], palette="cividis"
       ,saturation=5)
```

### CHANGE DEFAULT AGGREGATION METHOD USING ESTIMATOR PARAMETER

```python
plt.figure(figsize=(15, 5))

sns.barplot(x="Product line", y="Total", data=mart, order=['Electronic_accessories', 'Fashion_accessories',
       'Food_and_beverages', 'Health_and_beauty', 'Home_and_lifestyle',
       'Sports_and_travel'], palette="cividis"
       ,estimator=sum)
```
