### Import library

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### Read data

```python
mart = pd.read_excel("../Data/supermarket_sales.xlsx")

mart.head()
```

### Define which categorical plot we want to use

```python
sns.catplot(data=mart, x='Payment', y='Total') # Default to be strip plot
```

```python
sns.catplot(data=mart, x='Payment', y='Total', kind='box')
```

```python
sns.catplot(data=mart, x='Payment', y='Total', col='Gender', kind='violin', row='Customer type')
```

```python
sns.catplot(data=mart, x='Payment', y='Total', col='Gender', kind='violin', row='Customer type', palette='CMRmap_r')
```
